```python
%matplotlib inline
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
t=np.linspace(0,1,num=20000)
y=sig.chirp(t,0,1,10000)
f, t, Sxx = sig.spectrogram(y, len(y))
plt.xlabel('t')
plt.ylabel('f')
plt.pcolormesh(t, f, Sxx)
```

```python
y_new = np.zeros(10000)
for n in range (1, 10000):
    y_new[n] = y[2*n]
f, t, Sxx = sig.spectrogram(y_new, len(y_new))
plt.xlabel('t')
plt.ylabel('f')
plt.pcolormesh(t, f, Sxx)
```

Frequency higher than 5000 Hz grows in opposite direction due to aliasing

```python
y_new_2_de = np.zeros(10000)
H = np.zeros(20000)
for i in range (0, 10000):
    H[i] = 1
y_new_2 = H * y
for n in range (0, 10000):
    y_new_2_de[n] = y_new_2[2*n]
f, t, Sxx = sig.spectrogram(y_new_2_de, len(y_new_2_de))
plt.pcolormesh(t, f, Sxx)
```

The incorrect part is filtered out

```python
%matplotlib inline
import numpy as np
import numpy.fft as fft
import matplotlib.pyplot as plt

fs = 32000
fc = 8000
t = np.linspace(0, 1., fs)
car_wave = np.cos(2*np.pi*fc*t)
f_k = np.linspace(-fs/2, fs/2, fs)
f_k_2 = np.linspace(-fs/2, fs/2, 1024)
N = 1024
spectrum = fft.fftshift(np.abs(fft.fft(car_wave, N)))
plt.plot(f_k_2,spectrum)
plt.xlabel('f')
plt.ylabel(spectrum)
plt.show()
```

```python
f_new = 1000
mod_wave = np.cos(2 * np.pi * f_new * t)
mod_sig = (mod_wave/2 + 1) * car_wave
plt.subplot(1, 2, 1)
plt.plot(t[0:100], car_wave[0:100])
plt.xlabel('t')
plt.ylabel('carrier wave')
plt.subplot(1, 2, 2)
plt.plot(t[0:100], mod_sig[0:100])
plt.xlabel('t')
plt.ylabel('modulated signal')
plt.show()
```

```python
mod_sig_spec = fft.fftshift(np.abs(fft.fft(mod_sig, N)))
plt.subplot(1, 2, 1)
plt.plot(f_k_2, spectrum)
plt.xlabel('f')
plt.ylabel('carrier wave')
plt.subplot(1, 2, 2)
plt.plot(f_k_2, mod_sig_spec)
plt.xlabel('f')
plt.ylabel('modulated signal')
plt.show()
```
