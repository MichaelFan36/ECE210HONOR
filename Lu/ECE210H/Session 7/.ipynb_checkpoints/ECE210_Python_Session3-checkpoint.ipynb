##### Sampling Frequency

Bandwidth of Sine/Cosine Wave is just the Fundamental Frequency

```python
f=300.0;        # Frequency of Signal
fs_1=4000.0;    # Oversampling
fs_2=1000.0;    
fs_3=500.0;     # Undersampling

t1=np.arange(0.0,0.02,1/fs_1)
t2=np.arange(0.0,0.02,1/fs_2)
t3=np.arange(0.0,0.02,1/fs_3)
y1=np.sin(2*np.pi*f*t1); 
y2=np.sin(2*np.pi*f*t2);
y3=np.sin(2*np.pi*f*t3);

fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].plot(t1,y1,'r.-')
axes[0].set_xlabel('t')
axes[0].set_ylabel('y')
axes[1].plot(t2,y2,'b.-')
axes[1].set_xlabel('t')
axes[1].set_ylabel('y')
axes[2].plot(t3,y3,'g.-')
axes[2].set_xlabel('t')
axes[2].set_ylabel('y')

fig.tight_layout()
plt.show()
```

#### Autocorrelation between two signals

```python
y = np.sin(2*np.pi*np.arange(0, 1, 1/1000))
t = np.arange(0, 1, 1/1000)
fig, ax = plt.subplots(nrows=1, ncols=1)
y_corr = np.correlate(y,y,"full")
y_corr = y_corr[np.size(y_corr)/2]

ax.plot(t, y_corr, 'b--')
ax.set_xlabel('t')
ax.set_ylabel('y_corr(t)')
ax.set_title('output auto-correlation')

fig.tight_layout()
plt.show()
```

##### FFT Example for Sum of Cosines and DC gain

```python
fs=8000.0;
f_1=500.0; phase_1=0
f_2=750.0; phase_2=np.pi/4.0
f_3=1500.0; phase_3=np.pi/2.0
f_4=4500.0; phase_4=0; 
t=np.arange(0,1,1/fs)
y=1
y=y+np.cos(2*np.pi*f_1*t+phase_1)
y=y+0.5*np.cos(2*np.pi*f_2*t+phase_2)
y=y+np.cos(2*np.pi*f_3*t+phase_3)
y=y+np.cos(2*np.pi*f_4*t+phase_4)
plt.plot(t[0:80],y[0:80],'r.-') # Plot signal in time
plt.show()
```

```python
N=1024;
y_spec=fft.fft(y,N)
ff=np.linspace(-fs/2,fs/2,N) # Frequency axis
y_spec_mag=abs(fft.fftshift(y_spec)) #Plot signal in frequency domain
y_spec_phase=np.angle(fft.fftshift(y_spec))

fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(ff,y_spec_mag,'r')
axes.set_xlabel('frequency')
axes.set_ylabel('magnitude')

fig.tight_layout()
plt.show()
```

Reconstruct the signal using ifft. Signal must not be zero centered for iift.

```python
recon = np.fft.ifft(y_spec, N)
plt.figure(1)
plt.plot(t[0:80], abs(recon[0:80]))
```

```python
y_spec = np.fft.fftshift(y_spec) # zero centered
```

#### Exercise 1:

Create a rectangular filter of the same size as the signal y_spec above. The filter should be 1 between the frequencies -2000 and 2000. You will need to first create a vector of zeros and then find the indices that correspond to -2000 and 2000 and fill all array values between these two indices with 1. <br>
Call your filter H($\omega$). Then perform pointwise multiplication between the arrays H and y_spec. Then fftshift this multiplied signal and take the Inverse FFT using ifft. (Must fftshift because it is zero centered)<br>
Please note:<br>
y_spec is zero centered<br>
ff is the frequency axis you will be using in the plots<br>
Plots in the time domain should be done as in the first part of the above example.

```python
%matplotlib inline
# Create the filter. This is a zero centered filter.
H = np.zeros(1024)
# Fill in the correct indices with 1
for i in range (256, 768):
    H[i] = 1
# Then take the pointwise multiplication between h and y_spec. Plot the result. The frequency axis will be ff.
result = np.multiply(H, y_spec)
plt.plot(ff, abs(result))
plt.show()
# fftshift this product and then take the ifft. Plot your result in time, as in the time domain plot above.
shift = np.fft.fftshift(result)
inverse = np.fft.ifft(shift)
t = np.arange(0,1,1/8000)
plt.plot(t[0:80], np.abs(inverse[0:80]))
plt.show()
```

```python
# Take the ifft of the filter H and call it h
h = np.fft.ifft (H)
# Take the ifft of y_spec, This was already calculated as recon.
# Perform linear convolution
conv = np.convolve(h, np.fft.ifft(y_spec, N))

# Take the fft of this signal and plot its magnitude.
mag = np.fft.fft(conv, N)
plt.plot(ff, np.abs(mag))
plt.show()
```

#### (a.1) generate and plot the input signal x(t)

```python
#<------- space for you to code --------->
t = np.linspace(0, 10, 100)
f = 1/2 * np.cos(np.pi * t - np.pi/2)+4/(3*np.pi) * np.cos(np.pi/2 * t)
for n in range(3, 100, 2):
    f = f + 4/(np.pi*(n**2-4))*np.cos((n*np.pi*t)/2+np.pi)

plt.plot(t[0:100], f[0:100])
plt.xlabel('t')
plt.ylabel('x(t)')

# hint: have a variable x = x1 + x2 + ...
# hint: for-loop over n in range(3,100,2):
```

#### (a.2) write a function that returns the magnitude and phase of H(w) for given w

```python
# hint: replace all <text>
# hint: code this so that you can use the function in the next step as H_w_mag, H_w_phase = mag_phase_H(w) 

def function(H_w):
    H_w_mag = np.abs(H_w)
    H_w_phase = np.angle(H_w)
    return [H_w_mag, H_w_phase]
```

#### (a.3) plot magnitude and phase of H(w) against w

```python
w = np.linspace(-7*np.pi/4, 7*np.pi/4, 100)
H = np.pi/2 * (1+w**2)*np.e**(-1j*w)
H_w_mag, H_w_phase = function(H)

#<------- space for you to code --------->
plt.figure(1)
plt.plot(w[0:100], H_w_mag[0:100])
plt.xlabel('w')
plt.ylabel('H_w_mag')
plt.figure(2)
plt.plot(w[0:100], H_w_phase[0:100])
plt.xlabel('w')
plt.ylabel('H_w_phase')
```

#### (a.4) plot the steady-state response y_ss(t)

```python
t = np.linspace(0,10,100)
w = np.linspace(-7*np.pi/4, 7*np.pi/4, 100)

H_1 = np.pi/2 * (1+np.pi**2)*np.e**(-1.j*np.pi)
H_w_mag_1, H_w_phase_1 = function(H_1)
H_2 = np.pi/2 * (1+(np.pi/2)**2)*np.e**(-1.j*(np.pi/2))
H_w_mag_2, H_w_phase_2 = function(H_2)
y_ss = H_w_mag_1 * 1/2 * np.cos(np.pi * t - np.pi/2. + H_w_phase_1) + H_w_mag_2 * 4/(3*np.pi) * np.cos(np.pi/2 * t + H_w_phase_2)
# hint: sum up the terms (for loop not required because the range of H(w) is small)
# hint: use H_w_mag, H_w_phase = mag_phase_H(w) 
plt.plot(t, y_ss)
plt.xlabel('t')
plt.ylabel('y_ss')
```

#### (b) is y_ss(t) periodic? If so, obtain its fundamental frequency

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
y_corr = np.correlate(y_ss,y_ss,"full")
y_corr = y_corr[y_corr.size//2:]

ax.plot(t, y_corr, 'b--')
ax.set_xlabel('t')
ax.set_ylabel('y_corr(t)')
ax.set_title('output auto-correlation')

fig.tight_layout()
```
