##### Sampling Frequency

Bandwidth of Sine/Cosine Wave is just the Fundamental Frequency

```python
f=300.0;        # Frequency of Signal
fs_1=4000.0;    # Oversampling
fs_2=1500.0;    
fs_3=500.0;     # Undersampling

t1=np.arange(0.0,0.02,1/fs_1)
t2=np.arange(0.0,0.02,1/fs_2)
t3=np.arange(0.0,0.02,1/fs_3)
y1=np.sin(2*np.pi*f*t1); 
y2=np.sin(2*np.pi*f*t2);
y3=np.sin(2*np.pi*f*t3);

fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].plot(t1,y1,'r.-')
axes[0].set_xlabel('t')
axes[0].set_ylabel('y')
axes[0].set_title('Oversampled Signal')
axes[1].plot(t2,y2,'b.-')
axes[1].set_xlabel('t')
axes[1].set_ylabel('y')
axes[1].set_title('Just Correctly Sampled Signal')
axes[2].plot(t3,y3,'g.-')
axes[2].set_xlabel('t')
axes[2].set_ylabel('y')
axes[2].set_title('Undersampled Signal')

fig.tight_layout()
plt.show()
```

#### Autocorrelation of a signal

The autocorrelation of a signal $x$ is defined as:

\begin{equation}
a_x[n] = \sum_{m \in \mathbb{Z}} x[m]x[m-n]
\end{equation}

Notice that this is very similar to the definition of the discrete convolution:

\begin{equation}
(x \ast y)[n] = \sum_{m \in \mathbb{Z}} x[m]y[n-m]
\end{equation}

In fact, the autocorrelation of signal can be written in terms of a discrete convolution:

\begin{equation}
a_x[n] = \sum_{m \in \mathbb{Z}} x[m]x[m-n] = x[n] \ast x[-n]
\end{equation}

Basically, to compute the autocorrelation, take the convolution between the original signal and the reversed signal. Visually, you can just think of the autocorrelation as keeping one copy of the signal fixed while moving another copy of the signal over it and performed a dot product between overlapping elements.

The autocorrelation is used to find the time period of a signal in the example.

```python
y = np.sin(2*np.pi*np.arange(0, 2, 1/2000))
t = np.arange(0, 2, 1/2000)
fig, ax = plt.subplots(nrows=1, ncols=1)
# Autocorrelation computed using the correlate function
y_corr = np.correlate(y,y,"full")
# Autocorrelation using discrete convolution
# y_corr = np.convolve(y, y[::-1])

# Discard the first half since it will be symmetric to the second half
y_corr = y_corr[y_corr.size//2:]

ax.plot(t, y_corr, 'b--')
ax.set_xlabel('t')
ax.set_ylabel('y_corr(t)')
ax.set_title('output auto-correlation')

fig.tight_layout()
plt.show()
```

#### Cross Correlation between two signal

The cross correlation between two signals $x$ and $y$ is defined as:

\begin{equation}
a_xy[n] = \sum_{m \in \mathbb{Z}} x[m]y[m-n]
\end{equation}

Again this can be expressed as a convolution:

\begin{equation}
a_xy[n] = \sum_{m \in \mathbb{Z}} x[m]y[m-n] = x[n] \ast y[-n]
\end{equation}

Notice that the autotcorrelation is the cross correlation between the same signal! The cross correlation can be used to find traces of a hidden signal within a larger signal.

```python
t1 = np.arange(0, 4, 1/100)
t2 = np.arange(4, 8, 1/100)
y1 = np.concatenate((np.sin(2*np.pi*t1),0.5*np.sin(2*np.pi*0.5*t2)))

plt.figure()
plt.plot(np.concatenate((t1, t2)), y1)
plt.title('Large Signal')

hidden = np.sin(2*np.pi*0.5*t2)
xcorr = np.convolve(y1, hidden[::-1],"same")

plt.figure()
plt.plot(np.concatenate((t1, t2)),xcorr)
plt.title('Cross correlation')
plt.show()
```

##### FFT Example for Sum of Cosines and DC gain

```python
fs=8000.0;
f_1=500.0; phase_1=0
f_2=750.0; phase_2=np.pi/4.0
f_3=1500.0; phase_3=np.pi/2.0
f_4=4500.0; phase_4=0; 
t=np.arange(0,1,1/fs)
y=1
y=y+np.cos(2*np.pi*f_1*t+phase_1)
y=y+0.5*np.cos(2*np.pi*f_2*t+phase_2)
y=y+np.cos(2*np.pi*f_3*t+phase_3)
y=y+np.cos(2*np.pi*f_4*t+phase_4)
plt.plot(t[0:80],y[0:80],'r.-') # Plot signal in time
plt.title('Input signal')
plt.show()
```

```python
N=1024;
y_spec=fft.fft(y,N)
ff=np.linspace(-fs/2,fs/2,N) # Frequency axis
y_spec_mag=abs(fft.fftshift(y_spec)) #Plot signal in frequency domain
y_spec_phase=np.angle(fft.fftshift(y_spec))

fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(ff,y_spec_mag,'r')
axes.set_xlabel('frequency')
axes.set_ylabel('magnitude')
axes.set_title(r'DTFT of signal')

fig.tight_layout()
plt.show()
```

Reconstruct the signal using ifft. Signal must not be zero centered for iift.

```python
recon = np.fft.ifft(y_spec, N)
plt.figure(1)
plt.plot(t[0:80], np.real(recon[0:80]))
plt.title(r'Reconstructed Signal')
plt.show()
```
