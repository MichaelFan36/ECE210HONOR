```python
%matplotlib inline
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
t=np.linspace(0,1,num=20000)
y=sig.chirp(t,0,1,10000)
f, t, Sxx = sig.spectrogram(y, len(y))
plt.xlabel('t')
plt.ylabel('f')
plt.pcolormesh(t, f, Sxx)
```

```python
new_y = np.zeros(10000)
for n in range (1, 10000):
    new_y[n] = y[2*n]
f, t, Sxx = sig.spectrogram(new_y, len(new_y))
plt.xlabel('t')
plt.ylabel('f')
plt.pcolormesh(t, f, Sxx)
```

```python
#It's in the opposite direction since it's over 5000 Hz which is half of 10000Hz
```

```python
new_y2_d = np.zeros(10000)
H = np.zeros(20000)
for i in range (0, 10000):
    H[i] = 1
new_y2 = H * y
for n in range (0, 10000):
    new_y2_d[n] = new_y2[2*n]
f, t, Sxx = sig.spectrogram(new_y2_d, len(new_y2_d))
plt.pcolormesh(t, f, Sxx)
```

```python
%matplotlib inline
import numpy as np
import numpy.fft as fft
import matplotlib.pyplot as plt

fs = 32000
fc = 8000
number = 1024
t = np.linspace(0, 1., fs)
carrier_wave = np.cos(2*np.pi*fc*t)
f_k = np.linspace(-fs/2, fs/2, fs)
f_k_2 = np.linspace(-fs/2, fs/2, 1024)

spectrum = fft.fftshift(np.abs(fft.fft(carrier_wave, number)))
plt.plot(f_k_2,spectrum)
plt.xlabel('f')
plt.ylabel(spectrum)
plt.show()
```

```python
f_new = 1000
mod_wave = np.cos(2 * np.pi * f_new * t)
mod_sig = (mod_wave/2 + 1) * carrier_wave
plt.subplot(1, 2, 1)
plt.plot(t[0:100], carrier_wave[0:100])
plt.xlabel('t')
plt.ylabel('carrier wave')
plt.subplot(1, 2, 2)
plt.plot(t[0:100], mod_sig[0:100])
plt.xlabel('t')
plt.ylabel('modulated signal')
plt.show()
```

```python
mod_sig_spec = fft.fftshift(np.abs(fft.fft(mod_sig, number)))
plt.subplot(1, 2, 1)
plt.plot(f_k_2, spectrum)
plt.xlabel('f')
plt.ylabel('carrier wave')
plt.subplot(1, 2, 2)
plt.plot(f_k_2, mod_sig_spec)
plt.xlabel('f')
plt.ylabel('modulated signal')
plt.show()
```
